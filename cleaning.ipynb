```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,6))
```

```python
swig_df.boxplot(column="Ksat", by="Texture Class", grid=False, rot=45)
plt.title("Variation of Ksat within Texture Classes")
plt.suptitle("")
plt.ylabel("Ksat (cm h⁻¹)")
plt.xlabel("Texture Class")
plt.show()
```

```python
# 1. Remove records where Ksat > 1000 and Texture Class = "Clay"
swig_df = swig_df[~((swig_df["Texture Class"] == "CLAY") & (swig_df["Ksat"] > 1000))]

# 2. Remove record with maximum Ksat for "Loamy sand"
max_ksat_loamy_sand = swig_df.loc[swig_df["Texture Class"] == "LOAMY SAND", "Ksat"].idxmax()
swig_df = swig_df.drop(max_ksat_loamy_sand)

# 3. Remove record with maximum Ksat for "Silt clay loam"
max_ksat_silt_clay_loam = swig_df.loc[swig_df["Texture Class"] == "SILTY CLAY LOAM", "Ksat"].idxmax()
swig_df = swig_df.drop(max_ksat_silt_clay_loam)
```

```python
plt.figure(figsize=(12,6))
swig_df.boxplot(column="Ksat", by="Texture Class", grid=False, rot=45)
plt.title("Variation of Ksat within Texture Classes (Cleaned Data)")
plt.suptitle("")
plt.ylabel("Ksat (cm h⁻¹)")
plt.xlabel("Texture Class")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18,6), sharey=True)

# Helper function to add scatter + regression line
def scatter_with_line(ax, x, y, xlabel, title):
    ax.scatter(x, y, alpha=0.6)
    
    # Fit regression line
    coeffs = np.polyfit(x, y, 1)
    poly = np.poly1d(coeffs)
    xs = np.linspace(x.min(), x.max(), 100)
    ax.plot(xs, poly(xs), color="red")
    
    ax.set_xlabel(xlabel)
    ax.set_title(title)

# Clay vs Ksat
scatter_with_line(axes[0], swig_df["Clay"], swig_df["Ksat"], "Clay (%)", "Clay vs Ksat")
axes[0].set_ylabel("Ksat (cm h⁻¹)")

# Sand vs Ksat
scatter_with_line(axes[1], swig_df["Sand"], swig_df["Ksat"], "Sand (%)", "Sand vs Ksat")

# Silt vs Ksat
scatter_with_line(axes[2], swig_df["Silt"], swig_df["Ksat"], "Silt (%)", "Silt vs Ksat")

plt.tight_layout()
plt.show()
```
